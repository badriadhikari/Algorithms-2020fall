## Create graphs

```
import networkx as nx

G = nx.Graph()        # Undirected graph
G = nx.MultiGraph()   # Allows multiple undirected edges

G = nx.DiGraph()      # Directed graph
G = nx.MultiDiGraph() # Allows multiple directed edges
```

```
G = nx.Graph() 

# Add egdes
G.add_edge(1, 2)
G.add_edge(2, 3, weight = 0.9) # edge data is optional

# Add edges from a list
elist = [(1, 2), (2, 3), (1, 4), (4, 2)]
G.add_edges_from(elist)

G = nx.Graph() 
# Add weighted edges
elist = [('a', 'b', 5.0), ('b', 'c', 3.0), ('a', 'c', 1.0), ('c', 'd', 7.3)]
G.add_weighted_edges_from(elist)
```

## Plot graphs

```
# default spring_layout, with an existing example
import matplotlib.pyplot as plt
G = nx.cubical_graph() # get an example graph
plt.subplot(121)
nx.draw(G)
```

```
# default spring_layout, my own graph
import matplotlib.pyplot as plt
G = nx.Graph() 
elist = [('a', 'b', 5.0), ('b', 'c', 3.0), ('a', 'c', 1.0), ('c', 'd', 7.3), ('a', 'd', 5.0)]
G.add_weighted_edges_from(elist)
nx.draw(G, with_labels = True)
```

```
# circular layout
import matplotlib.pyplot as plt
G = nx.cubical_graph()
plt.subplot(122)
nx.draw(G, pos=nx.circular_layout(G), node_color='r', edge_color='b')
```

```
# circular layout, my own graph
import matplotlib.pyplot as plt
G = nx.Graph() 
elist = [('a', 'b', 5.0), ('b', 'c', 3.0), ('a', 'c', 1.0), ('c', 'd', 7.3), ('a', 'd', 5.0)]
G.add_weighted_edges_from(elist)
plt.subplot(122)
nx.draw(G, pos=nx.circular_layout(G), node_color='r', edge_color='b', with_labels = True)
```
